### Recommendation to Improve Bank Account Ownership

Based on the analysis, bank account ownership is strongly influenced by education, employment type, cellphone access, and location. To improve financial inclusion, the following actions are recommended:

####  1. Improve digital access

People without cellphones are far less likely to have bank accounts. Expanding affordable mobile phone access and digital literacy programs can significantly increase financial inclusion.

####  2. Focus on rural populations

Rural residents have lower bank account ownership than urban residents. Expanding banking services and mobile banking agents in rural areas can reduce this gap.

####  3. Support low-education groups

Individuals with no or low education levels are much less likely to have bank accounts. Financial education programs should target these groups to improve awareness and trust in banking services.

####  4. Promote informal sector inclusion

People in informal jobs, farming, and fishing have very low bank account usage. Tailored financial products (simple accounts, mobile banking) should be designed for informal workers.

####  5. Encourage formal employment pathways

Formal employment is strongly linked to higher bank account ownership. Policies that support job formalization can indirectly improve financial inclusion.

####  6. Address gender gap

Females are less likely to have bank accounts than males. Targeted programs for women, such as financial literacy and women-focused banking services, are needed.

####  7. Country-level targeting

Some countries show lower inclusion rates. Country-specific financial inclusion strategies should be developed rather than using a one-size-fits-all approach.

### New Feature Creation

#### for train

```python
#  Education strength score 
train['edu_strength'] = (
   
    1 * train['education_level_Primary education'] +
    2 * train['education_level_Secondary education'] +
    3 * train['education_level_Vocational/Specialised training'] +
    4 * train['education_level_Tertiary education']
)

#  High education indicator 
train['high_edu'] = (
    train['education_level_Tertiary education'] |
    train['education_level_Vocational/Specialised training']
).astype(int)

#  Formal job indicator 
train['formal_job'] = (
    train['job_type_Formally employed Government'] |
    train['job_type_Formally employed Private']
).astype(int)

#  Economic stability index 
train['economic_stability'] = (
    train['high_edu'] +
    train['formal_job'] +
    train['cellphone_access_Yes']
)

#  Exclusion risk 
train['exclusion_risk'] = (
    train['education_level_Primary education'] +
    train['job_type_No Income'] +
    train['cellphone_access_Yes']
)

#  Interaction feature 
train['edu_job_formal'] = (
    train['education_level_Secondary education'] *
    train['job_type_Formally employed Private']
)
```

#### for test

```python
#  Education strength score 
test['edu_strength'] = (
   
    1 * test['education_level_Primary education'] +
    2 * test['education_level_Secondary education'] +
    3 * test['education_level_Vocational/Specialised training'] +
    4 * test['education_level_Tertiary education']
)

#  High education indicator 
test['high_edu'] = (
    test['education_level_Tertiary education'] |
    test['education_level_Vocational/Specialised training']
).astype(int)

#  Formal job indicator 
test['formal_job'] = (
    test['job_type_Formally employed Government'] |
    test['job_type_Formally employed Private']
).astype(int)

#  Economic stability index 
test['economic_stability'] = (
    test['high_edu'] +
    test['formal_job'] +
    test['cellphone_access_Yes']
)

#  Exclusion risk 
test['exclusion_risk'] = (
    test['education_level_Primary education'] +
    test['job_type_No Income'] +
    test['cellphone_access_Yes']
)

#  Interaction feature 
test['edu_job_formal'] = (
    test['education_level_Secondary education'] *
    test['job_type_Formally employed Private']
)
```

```python
selected_features=['year', 'household_size',
       'age_of_respondent', 'age_group', 'household_size_group',
       'country_Rwanda', 'country_Tanzania', 'country_Uganda',
       'location_type_Urban', 'cellphone_access_Yes',
       'gender_of_respondent_Male', 'relationship_with_head_Head of Household',
       'relationship_with_head_Other non-relatives',
       'relationship_with_head_Other relative',
       'relationship_with_head_Parent', 'relationship_with_head_Spouse',
       'marital_status_Dont know', 'marital_status_Married/Living together',
       'marital_status_Single/Never Married', 'marital_status_Widowed',
       'education_level_Other/Dont know/RTA',
       'education_level_Primary education',
       'education_level_Secondary education',
       'education_level_Tertiary education',
       'education_level_Vocational/Specialised training',
       'job_type_Farming and Fishing', 'job_type_Formally employed Government',
       'job_type_Formally employed Private', 'job_type_Government Dependent',
       'job_type_Informally employed', 'job_type_No Income',
       'job_type_Other Income', 'job_type_Remittance Dependent',
       'job_type_Self employed', 'high_edu', 'edu_strength', 'formal_job',
       'economic_stability', 'exclusion_risk', 'edu_job_formal']
```

```python
x=train[selected_features]
y=train[ 'bank_account']
x_test=test[selected_features]
x_train, x_val, y_train, y_val = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.utils.class_weight import compute_class_weight
classes = np.unique(y_train)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = dict(zip(classes, weights))

neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
scale_pos_weight = neg / pos
```

```python
models = {
    "Logistic Regression": Pipeline([

        ('model', LogisticRegression(max_iter=1000, class_weight='balanced'))
    ]),
    
    "Random Forest": Pipeline([
 
        ('model', RandomForestClassifier(
            n_estimators=200,
            random_state=42,
            class_weight='balanced'
        ))
    ]),
    
    "XGBoost": Pipeline([
     
        ('model', xgb.XGBClassifier(
            scale_pos_weight=scale_pos_weight,
            eval_metric='logloss',
            use_label_encoder=False
        ))
    ]),
    
    "LightGBM": Pipeline([
   
        ('model', lgb.LGBMClassifier(
            class_weight='balanced'
        ))
    ])
}
```

```python
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score, classification_report
```

```python
results = []

for name, model in models.items():
    model.fit(x_train, y_train)
    
    preds = model.predict(x_val)
    probs = model.predict_proba(x_val)[:, 1]
    
    acc = accuracy_score(y_val, preds)
    auc = roc_auc_score(y_val, probs)
    f1 = f1_score(y_val, preds)
    
    print(f"\n{name}")
    print(classification_report(y_val, preds))
    
    results.append({
        "Model": name,
        "Accuracy": acc,
        "ROC_AUC": auc,
        "F1": f1
    })

results_df = pd.DataFrame(results).sort_values(by="F1", ascending=False)
print("\nFinal Comparison (sorted by F1):")
print(results_df)
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('pass', 'passthrough', x_train.columns)
    ]
)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', xgb.XGBClassifier(
        scale_pos_weight=scale_pos_weight,
        eval_metric='logloss'
    ))
])
```

```python
param_grid = {
    'model__n_estimators': [200, 300, 500],
    'model__max_depth': [3, 5, 7],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__subsample': [0.7, 0.8, 1.0],
    'model__colsample_bytree': [0.7, 0.8, 1.0],
    'model__min_child_weight': [1, 3, 5]
}

grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='f1',
    cv=3,
    verbose=2,
    n_jobs=-1
)
grid_search.fit(x_train, y_train)
best_model = grid_search.best_estimator_
print(grid_search.best_params_)
```

```python
probs_val = best_model.predict_proba(x_val)[:,1]
best_f1 = 0
best_thresh = 0
for t in np.linspace(0.1, 0.9, 50):
    preds = (probs_val > t).astype(int)
    score = f1_score(y_val, preds)
    
    if score > best_f1:
        best_f1 = score
        best_thresh = t

print("Best Threshold:", best_thresh)
print("Best F1 (val):", best_f1)
```

```python
probs_test = best_model.predict_proba(x_test)[:,1]
preds_test = (probs_test > best_thresh).astype(int)
preds_test
```
